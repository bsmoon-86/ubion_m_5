```python
# 라이브러리 로드 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## matplotlib을 이용한 그래프의 시각화 
# 라인 그래프 (플롯 그래프)
# y축의 데이터만 존재하는 경우
# x축의 데이터는 0부터 시작해서 1씩 증가하는 형태로 생성
y_list = [10, 5, 8, 16]
plt.plot(y_list)
# show() : print() 같은 기능
plt.show()
```

```python
# x, y축의 데이터가 모두 존재하는 경우 
x_list = [1, 3, 6, 10]
y_list = [1,2,3,4]
plt.plot(x_list, y_list)
plt.show()
```

```python
# 라인 그래프의 라인의 색상과 종류를 변경
plt.plot(x_list, y_list, '#FF0077')
plt.show()
```

```python
plt.plot(x_list, y_list, '--')
plt.show()
```

```python
plt.plot([1,2,3,4], [10, 20, 10, 30])
plt.plot([1,3,5,7], [20, 40, 8, 1])
# x축에 라벨링 
plt.xlabel('X')
# y축 라벨링
plt.ylabel('Y')
# 그래프의 이름 표시 
plt.title('그래프')
# 범례 추가 
plt.legend(['a', 'b'])
plt.show()
```

```python
df = pd.read_csv("../../csv/drinks.csv")
```

```python
df.head()
```

```python
# 맥주소비량, 위스키소비량, 와인의 소비량, 순수 알콜의 양의 
# 상관 관계 확인
# 상관 계수 : -1 ~ 1
    # -1 : 하나의 데이터가 오르면 나머지 하나의 데이터는 내려간다. 
    # 1 : 하나의 데이터가 오르면 나머지 하나의 데이터도 같이 오른다. 
    # 0 : 두 개의 데이터가 무작위
# DataFrame 안에 있는 corr() 함수 호출
corr = df[ ['beer_servings', 'spirit_servings', 'wine_servings',
     'total_litres_of_pure_alcohol'] ].corr('pearson')
```

```python
# seaborn 라이브러리를 이용하여 히트맵 시각화
cols = ['beer', 'spirit', 'wine', 'alcohol']

hm = sns.heatmap(
    corr.values, 
    cbar = True,    # 사이드바 표시 
    annot = True,   # 히트맵에 숫자를 표시 
    square= True,   # 정사각형 표시
    fmt = '0.3f',   # 숫자의 소수점 자리수
    annot_kws = {'size' : 20}, # 숫자의 크기 
    xticklabels= cols, 
    yticklabels= cols
)

plt.show()
```

```python
sns.pairplot(
    df.loc[:, 'beer_servings' : 'total_litres_of_pure_alcohol'], 
    height = 2.5, 
    kind = 'reg'
)
plt.show()
```
